```python
df = df.drop('Unnamed: 25', axis=1)  #nos permite quitarle una columna al dataframe, axis indica si es fila o columna lo que quiero eliminar, en las columnas siempre es 1, en filas es 0. 
```

```python
df = df[["MONTH", "DAY_OF_MONTH", "DAY_OF_WEEK", "ORIGIN", "DEST", "CRS_DEP_TIME", "ARR_DEL15"]]     #filtro las columnas que yo quiero tener, no todas son necesarias asÃ­ que las filtro. es obligatorio meterlo en dos corchetes
df.isnull().sum()
```

```python
#vemos que tenemos valores Nan, entonces procedemos a cambiar ese valor para poder facilitar el analisis
df = df.fillna({'ARR_DEL15':1}) #rellena los datos Nan con uno dentro de la columna ARR_DEL15 con un UNO
df.iloc[177:185]     #de todo el dataframe imprime las filas de la 177 al 184, el ultimo nunca lo toma en cuenta
```

```python
df.head()
```

```python
#los modelos de machine learning no aceptan letras, solo analizan numeros
#por lo anterior creamos DUMMIES entonces
import math

for index, row in df.iterrows():    #iterrows nos permite crear un ciclo que nos permite manejar fila por fila
    df.loc[index, 'CRS_DEP_TIME'] = math.floor(row['CRS_DEP_TIME'] / 100)   #leemos fila por fila de la columna CRS_DEP_TIME y le aplicamos floor que es el redondear los valores, entonces a esa fila la vas a redondear y la vas a dividir entre 100, esto es para que divida el tiempo de salida de CRSDEPTIME para facilitar la informacion 
df.head()
```

```python
#observamos el cambio en la columna CRS_DEP_TIME
#ahora si creamos los dummies
df = pd.get_dummies(df, columns=['ORIGIN', 'DEST']) # aplicacamos la funcion getdummies a dos columnas la de ORIGIN y la de DESTINO 
df.head()
```

# Entrenamos el modelo

```python
#dividimos el set frame en otros set frames
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df.drop('ARR_DEL15', axis=1), df['ARR_DEL15'], test_size=0.2, random_state=42)  #test size es para decirle cuanto % va agarrar de todo el data frame para entrenar, en este caso es el 20, random state es para predecir los resultados. 
```

```python
train_x.shape   #el 20% que tomo de entrenamiento
```

```python
test_x.shape    #el 80% de prueba que tomo del dataframe
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=13)
model.fit(train_x, train_y) 
```

```python
#ahora vamos a evaluar el modelo
predicted = model.predict(test_x)   #le decimos que prediga con las variables, nos va a dar todas las predicciones del model
model.score(test_x, test_y)     #evalua el modelo en las variables (testx) contra los resultados reales (testy) nos va imprimir la precision del modelo
```

```python
from sklearn.metrics import roc_auc_score
probabilities = model.predict_proba(test_x)
```

```python
roc_auc_score(test_y,probabilities[:,1])    #nos da el porcentaje de probabilidad de llegar tarde en el primer vuelo, nos imprime el area bajo la curva
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(test_y, predicted)   
```

```python
from sklearn.metrics import precision_score

train_predictions = model.predict(train_x)

precision_score(train_y, train_predictions)
```

```python
from sklearn.metrics import recall_score
recall_score(train_y, train_predictions)
```

```python
def predict_delay(departure_date_time, origin, destination):
    from datetime import datetime

 

    try:
        departure_date_time_parsed = datetime.strptime(departure_date_time, '%d/%m/%Y %H:%M:%S')
    except ValueError as e:
        return 'Error parsing date/time - {}'.format(e)

 

    month = departure_date_time_parsed.month
    day = departure_date_time_parsed.day
    day_of_week = departure_date_time_parsed.isoweekday()
    hour = departure_date_time_parsed.hour

 

    origin = origin.upper()
    destination = destination.upper()

 

    input = [{'MONTH': month,
              'DAY': day,
              'DAY_OF_WEEK': day_of_week,
              'CRS_DEP_TIME': hour,
              'ORIGIN_ATL': 1 if origin == 'ATL' else 0,
              'ORIGIN_DTW': 1 if origin == 'DTW' else 0,
              'ORIGIN_JFK': 1 if origin == 'JFK' else 0,
              'ORIGIN_MSP': 1 if origin == 'MSP' else 0,
              'ORIGIN_SEA': 1 if origin == 'SEA' else 0,
              'DEST_ATL': 1 if destination == 'ATL' else 0,
              'DEST_DTW': 1 if destination == 'DTW' else 0,
              'DEST_JFK': 1 if destination == 'JFK' else 0,
              'DEST_MSP': 1 if destination == 'MSP' else 0,
              'DEST_SEA': 1 if destination == 'SEA' else 0 }]

 

    return model.predict_proba(pd.DataFrame(input))[0][0]
```
